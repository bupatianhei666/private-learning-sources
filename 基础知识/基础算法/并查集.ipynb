# 并查集  
主要进行区间合并和查找两个点是否同属于一个区间  
## acwing并查集  
   
```c
#include <iostream>

using namespace std;
const int N = 100010;
int n, m, p[N];

int find(int x)
{
    //并查集合并操作
    if(x != p[x]) p[x] = find(p[x]);
    return p[x];
}

int main()
{
    cin >> n >> m;

    //一共有n个数，各自组成一个集合
    for(int i = 1;i <= n;i++) p[i] = i;

    while(m--)
    {
        char op[2];
        cin >> op;
        int a, b;
        if(op[0] == 'M')
        {
            //集合合并
            cin >> a >> b;
            if(!(find(a) == find(b))) p[find(a)] = find(b);
        }else
        {
            cin >> a >> b;
            if(find(a) == find(b)) cout << "Yes" << endl;
            else
            cout << "No" << endl;
        }
    }

    return 0;
}
```
