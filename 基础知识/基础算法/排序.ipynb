# 快速排序  
## acwing模板  
```c
#include <iostream>
#include <random>
#include <ctime>
using namespace std;
const int N = 100010;
int a[N], n;

void quick_sort(int l, int r)
{
    if(l >= r)
    return;
    //设置随机数种子
    srand(time(0));
    //设置基准值的位置
    int t = rand()%(r - l + 1) + l;
    int q = a[t], ll = l - 1, rr = r + 1;
    while(ll < rr)
    {
        do ll++;while(a[ll] < q);
        do rr--;while(a[rr] > q);
        if(ll < rr)
        swap(a[ll], a[rr]);
    }
    quick_sort(l, rr);
    quick_sort(rr + 1, r);
}

int main()
{
    cin >> n;
    for(int i = 0;i < n;i++)
    cin >> a[i];

    quick_sort(0, n - 1);

    for(int i = 0;i < n;i++)
    cout << a[i] << " ";
    cout << endl;
    return 0;
}
```
# 归并排序  
```c
#include <iostream>

using namespace std;
const int N = 100010;
int n, a[N], b[N];

void merge_sort(int l, int r)
{
    if(l >= r)
    return;
    int mid = l + r >> 1;
    merge_sort(l, mid);
    merge_sort(mid + 1, r);
    int ll = l, rr = mid + 1, k = 0;
    while(ll <= mid && rr <= r)
    {
        if(a[ll] < a[rr])
        b[k++] = a[ll++];
        else
        b[k++] = a[rr++];
    }
    while(ll <= mid)
    b[k++] = a[ll++];
    while(rr <= r)
    b[k++] = a[rr++];
    for(int i = l, j = 0;j <k;j++, i++)
    {
        a[i] = b[j];
    }
}

int main()
{
    cin >> n;
    for(int i = 0;i < n;i++) cin >> a[i];

    merge_sort(0, n - 1);

    for(int i = 0 ;i < n;i++)
    cout << a[i] << " ";
    cout << endl;
    return 0;
}
```
# 堆排序  
```c
#include <iostream>

using namespace std;
const int N = 100010;
int n, m, a[N];

void down(int k, int size)
{
    int t = k;
    if(2*k <= size && a[2*k] < a[t]) t = 2*k;
    if(2*k + 1 <= size && a[2*k + 1] < a[t]) t = 2*k + 1;
    if(t != k)
    {
        swap(a[k], a[t]);
        down(t, size);
    }
}

int main()
{
    cin >> n >> m;
    for(int i = 1;i <= n;i++) cin >> a[i];

    //进行down操作，构建堆，从n/2处开始执行down操作，构造一个小顶堆
    for(int i = n/2;i >= 1;i--)
    {
        down(i, n);
    }

    for(int i = 0;i < m;i++)
    {
        cout << a[1] << " ";
        swap(a[1], a[n--]);
        down(1, n);
    }
    return 0;
}
```
