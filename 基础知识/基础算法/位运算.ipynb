# 统计二进制中1的个数
- **x&(-x)得到二进制中最后一个1及以后的二进制所表示的值** 
- **x&(x - 1)则可以消去该数的二进制表示中最后一个1以及以后的值**
### 方法一
```c
int countT(int x)
{
    int ans = 0;
    while(x > 0)
    {
        if(x & 1 == 1)
        ans++;
        x = x >> 1;
    }
    return ans;
}
```
### 方法二
```c
int countT(int x)
{
    int ans = 0;
    while(x > 0)
    {
        //消去二进制中的最后一位1
        x = x&(x - 1);
        ans++;
    }
    return ans;
}
```
### 方法三
```c
int countT(int x)
{
    int ans = 0;
    while(x > 0)
    {
        // 利用x&-x得到二进制中最后一位1及以后的值
        int t = x&(-x);
        x -= t;
        ans++;
    }
    return ans;
}
```
# 输出一个数的二进制表示
```c
int main()
{
    int x;
    cin >> x;
    int flag = 0;
    // 输出二进制数，并去除前导0
    for(int i = 31;i >= 0;i--)
    {
        int t = x >> i & 1;
        if(t)
        {
            flag = 1;
        }
        if(flag)
        {
            cout << t;
        }
    }
}
```
