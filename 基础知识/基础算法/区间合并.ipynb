# 区间合并  
acwing区间合并题目  
  
### 代码如下  
```c
#include <iostream>
#include <vector>
#include <algorithm>

using namespace std;
typedef pair<int, int> PII;

int n;
vector<PII> a;

void merge()
{
    vector<PII> ans;
    int st = -2e9, ed = -2e9;
    for(PII item: a)
    {
        //上一个区间的尾部比较当前区间的头部，判断是否相交，如果相交则进行合并
        if(ed < item.first)
        {
            //区间不相交
            if(st != -2e9)
            ans.push_back({st, ed});
            //重新设置比较区间的开头和结尾
            st = item.first;
        }
        //相交的话，比较哪个区间的结尾更大，进行更新
        ed = max(ed, item.second);
    }
    //将最后一个区间计入答案，同时为了防止给定数组为空，判断st是否为-2e9
    if(st != -2e9)
    ans.push_back({st, ed});
    a = ans;
}

int main()
{
    scanf("%d", &n);
    for(int i = 0;i < n;i++)
    {
        int l, r;
        scanf("%d%d", &l, &r);
        a.push_back({l, r});
    }

    //按照区间的开头对区间进行排序
    sort(a.begin(), a.end());

    //区间合并
    merge();

    cout << a.size() << endl;
    return 0;
}
```
