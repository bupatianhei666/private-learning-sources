```c
#include <iostream>
#include <vector>
#include <algorithm>

using namespace std;
typedef pair<int, int> PII;

int n;
vector<PII> a;

void merge()
{
    vector<PII> ans;
    int st = -2e9, ed = -2e9;
    for(PII item: a)
    {
        //
        if(ed < item.first)
        {
            //
            if(st != -2e9)
            ans.push_back({st, ed});
            //
            st = item.first;
        }
        //
        ed = max(ed, item.second);
    }
    //st-2e9
    if(st != -2e9)
    ans.push_back({st, ed});
    a = ans;
}

int main()
{
    scanf("%d", &n);
    for(int i = 0;i < n;i++)
    {
        int l, r;
        scanf("%d%d", &l, &r);
        a.push_back({l, r});
    }

    //
    sort(a.begin(), a.end());

    //
    merge();

    cout << a.size() << endl;
    return 0;
}
```
