```python
import numpy as np
import pandas as pd

# pandas有两种数据结构，Series相当于一位数据结构，DataFrame相当于二维数据结构
d =pd.Series([1, 4, 5, 7], index = list('abcd'))
print('一维数据结构打印如下：')
print(d, type(d))

dic = {
    'country': ['china', 'china', 'china', 'japan', 'japan', 'japan'],
    'city': ['beijing', 'shanghai', 'guangzhou', 'tokyo', 'osaka', 'kyoto'],
    'year': [2000, 2001, 2002, 2001, 2002, 2003],
}
# 默认index从0开始，这里改成从1开始，column采用字典的键
df = pd.DataFrame(dic, index = range(1, 7))
print('二维数据结构打印如下：')
print(df)
```
