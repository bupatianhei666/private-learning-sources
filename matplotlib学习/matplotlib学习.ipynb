# 这是个matplotlib各种图的画法教程

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["font.sans-serif"] = ["SimSun"]
#  解决中文字体乱码问题
plt.rcParams["axes.unicode_minus"] = False
#  解决图像中的负号乱码问题

x = np.linspace(-10, 10, 1000)
y = (1/np.sqrt(2*np.pi))*np.exp(-np.square(x - 1)/2)
# lw设置线条粗细，ls设置线条样式
plt.plot(x, y, label='正态分布图像', lw = 3, ls = ':')
plt.xlabel("x")
plt.ylabel("y")
# 设置y轴显示范围 只管边界
plt.ylim(0, 0.5)
plt.xlim(-10, 10)
# 设置x轴的实际显示范围 可以设置字体大小，旋转角度
plt.xticks(np.linspace(-10, 10, 21, endpoint = True), fontsize = 10, rotation = 45)
# 给图像添加注释
# 在对应坐标加一个点
plt.plot([1], [0.4], 'o')
plt.annotate("极大值", xy = (1, 0.4), xytext = (1.2, 0.45), xycoords = 'data', textcoords = 'data',arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad = 0.2'))
plt.title("正态分布图像")
# 开启网格
plt.grid(True)
# 开启图画中小标题
plt.legend(loc = 'best')
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y1 = x**2
y2 = x**3
y3 = x
# 三幅子图
plt.figure(figsize=(5, 3))
plt.subplot(1, 3, 1)
plt.plot(x, y1, ls = '-', c = 'r')
plt.title('y = x^2')
plt.subplot(1, 3, 2)
plt.plot(x, y2, ls = '--', c = 'g')
plt.title('y = x^3')
plt.subplot(1, 3, 3)
plt.plot(x, y1, ls = '-.')
plt.title('y = x')
# 调整三幅图之间的距离 水平距离为wspace,垂直距离为hspace
plt.subplots_adjust(wspace = 0.5)
plt.show()
```

```python
# 画一个直方图
x = [12,  23, 34, 45, 56, 67, 78, 89, 90, 99]
plt.figure(figsize=(5, 3))
plt.subplot(1, 2, 1)
# bins设置长条形数目，facecolor设置长条形颜色，edgecolor设置边框颜色
plt.hist(x, bins=20, facecolor="green", edgecolor="black", linewidth=1.2)
plt.subplot(1, 2, 2)
# histtype = step设置柱状颜色不填充
plt.hist(x, bins=30, histtype="step", facecolor="green", edgecolor="black", linewidth=1.2)
plt.show()
```

```python
# 画一个柱状图
x = ['lisi', 'wangwu', 'liuli', 'zhansan', 'zhaoliu']
y = np.array([3, 8, 1, 10, 5])
plt.figure(figsize=(5, 3))
plt.bar(x, y, color="grey", width=0.5)
plt.show()
```

```python
labels = ['人事部', '技术部', '宣传部', '组织部']
men = [18, 19, 20, 21]
women = [23, 4, 13, 2]

x = np.arange(len(labels))
width = 0.2

fig, ax = plt.subplots()
# 设置图像大小
fig.set_size_inches(5, 3)
res1 = ax.bar(x - width/2, men, width, label = '男生人数')
res2 = ax.bar(x + width/2, women, width, label = '女生人数')
ax.set_xlabel('部门')
ax.set_ylabel('人数')
ax.set_title('分组柱状图')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(loc = 'best')
plt.show()
```

```python
x = np.arange(0, 1, 0.01)
y1 = np.random.rand(len(x))
y2 = np.random.rand(len(x))
# 同时画两幅图进行对比
plt.subplot(2, 1, 1)
plt.scatter(x, y1, alpha=0.3)
plt.title('plot 1')
plt.subplot(2, 1, 2)
plt.scatter(x, y2, alpha=1)
plt.title('plot 2')
# 调整两幅图之间的距离
plt.subplots_adjust(hspace=0.5)
# 调整两幅图的布局
plt.tight_layout()
plt.show()
```

```python
# 画一个饼状图
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
labels2 = 'Frogs', 'Hogs', 'Dogs', 'Logs'
size2 = [115, 30, 45, 10]
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0.05, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
plt.subplot(1, 2, 1)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.subplot(1, 2, 2)
plt.pie(size2, explode=explode, labels=labels2, colors=colors,
     autopct='%d%%',shadow=True, startangle=30)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# 制造输入的数据
square = [value ** 2 + 1 for value in np.arange(0, 12, 0.1)]
square1 = [value * 8 + 1 for value in np.arange(0, 12, 0.1)]
x = [value for value in np.arange(0, 12, 0.1)]

# 找到两个图像相交的部分的布尔数组
# 不能直接让where = square1 > square 因为列表不能进行比较作为条件
intersection_indices = np.array(square1) > np.array(square)
# 设置显示图像的大小
fig = plt.figure(dpi=128, figsize=(5, 4))
plt.plot(x, square)
plt.plot(x, square1)
# 在两条线相交的地方填充颜色
plt.fill_between(x, square, square1, where =intersection_indices,facecolor='blue', alpha=0.3)
plt.annotate('相交部分', xy=(4, 24), xytext=(6, 28), arrowprops=dict(facecolor='black', shrink=0.005))
plt.tick_params(axis='both', labelsize=10)
# 设置x的坐标为斜着的
fig.autofmt_xdate()

# 进行规则的颜色填充
plt.fill([10, 20, 30], [40, 70, 40], color = 'y')

plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos", color="red", linewidth=2)
plt.fill_between(x, y1, y2, where = y1 >= y2, color = 'grey')
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["font.sans-serif"] = ["SimSun"]
plt.rcParams["axes.unicode_minus"] = False

x = np.linspace(-10, 10, 100, endpoint=True)
y = np.linspace(-10, 10, 100, endpoint=True)
# 将数据进行网格化处理
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2+Y**2)

fig, (ax1, ax2) = plt.subplots(1, 2, subplot_kw={"projection": "3d"})
# rstride、cstride为横竖的采样步长，使图像更加细腻
ax1.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap="rainbow")
ax1.set(xlabel="x", ylabel="y", zlabel="z")

ax2.contourf(X, Y, Z, zdir="z", offset=0, cmap="rainbow")
ax2.set(xlabel="x", ylabel="y", zlabel="z")

# 设置子图间距
plt.subplots_adjust(wspace=0.4, hspace=0.4)
plt.show()
```
